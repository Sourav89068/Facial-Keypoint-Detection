```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
import glob
```

```python
glob.glob("adam_mse_lr50/*")
```

```python
a=os.listdir("adam_mse_lr50/")
```

```python
def func(dir):
    file_names=os.listdir(dir+"/")
    file_names=sorted(file_names)
    l=[]
    for i in file_names:
        a=pd.read_csv(dir+"/"+i)
        if i[7]==".":
            c=i[0:7]+"0"
            b={c:a}
        else:
            b={i[0:8]:a}
        l.append(b)
    return l
```

```python
files=func("adam_mse_lr50")
```

```python
def plot_loss(lis,ylim=(0,40)):
    plt.figure(figsize=(20,10))
    plt.subplot(121)
    for i in range(len(lis)):
        plt.plot(list(lis[i].values())[0]["train_loss"],label=list(lis[i].keys())[0]+" loss="+str(list(list(lis[i].values())[0]["train_loss"])[-1]))
        plt.title("Training loss",fontdict={'family': 'serif','color': 'Black','weight': 'normal','size': 30,})
    plt.legend()
    plt.ylim(ylim)
    plt.xlabel("Number of epochs",fontdict={'family': 'serif','color': 'darkBlue','weight': 'normal','size': 20,})
    plt.subplot(122)
    for j in range(len(lis)):
        plt.plot(list(lis[j].values())[0]["valid_loss"],label=list(lis[j].keys())[0]+" loss="+str(list(list(lis[j].values())[0]["valid_loss"])[-1]))
        plt.title("Validation loss",fontdict={'family': 'serif','color': 'Black','weight': 'normal','size': 30,})
    plt.legend()
    plt.xlabel("Number of epochs",fontdict={'family': 'serif','color': 'darkBlue','weight': 'normal','size': 20,})
    plt.ylim(ylim)
    return
```

# Losses with different learning rate in 50 epochs 

```python
plot_loss(files)
```

```python
plot_loss(files,(0,4))
```

# Losses with different learning rate in 100 epochs 

```python
files1=func("adam_mse_lr100/")
```

```python
plot_loss(files1)
```
