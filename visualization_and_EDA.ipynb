# Necessary imports 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Reading the data 

```python
train_data=pd.read_csv("training.csv")
```

# Displaying the ground truths 

```python
from groundtruth import show_images_with_keypoints
show_images_with_keypoints(train_data,np.random.randint(0,len(train_data),20))
```

# Calculating the missing values in ground truth

```python
a=train_data.isnull().sum()[0::2]
b=[a.index[i].replace("_x","").replace("_"," ").upper() for i in range(16)]
c=a.values
```

```python
plt.figure(figsize=(10,8))
sns.barplot(c,b,palette="tab20")
for index, value in enumerate(c):
    plt.text(value, index, str(value))
sns.set(rc={'figure.figsize':(11,5)}, font_scale=1.2)
plt.xlabel("Number of missing values",fontdict={'family': 'serif','color': 'darkmagenta','weight': 'normal','size': 16,})
plt.ylabel("Keypoints",fontdict={'family': 'serif','color': 'darkmagenta','weight': 'normal','size': 16,})
plt.title("Missing Labels",fontdict={'family': 'serif','color': 'darkBlue','weight': 'normal','size': 20,})
```

# Correlation matrices of x coordinates and y coordinates

```python
plt.figure(figsize=(20,8))
mask=np.triu(np.ones_like(train_data.iloc[:,0:30:2].corr(),dtype=np.bool))
heatmap=sns.heatmap(train_data.iloc[:,0:30:2].corr(),mask=mask,vmin=-1,vmax=1,annot=True,cmap='YlGnBu')
heatmap.set_title("Correlation matrix of x co-ordinates",fontdict={'family': 'serif','color': 'darkBlue','weight': 'normal','size': 20,},pad=12)
```

```python
plt.figure(figsize=(20,8))
mask=np.triu(np.ones_like(train_data.iloc[:,1:30:2].corr(),dtype=np.bool))
heatmap=sns.heatmap(train_data.iloc[:,1:30:2].corr(),mask=mask,vmin=-1,vmax=1,annot=True,cmap="YlGnBu")
heatmap.set_title("Correlation matrix of y co-ordinates",fontdict={'family': 'serif','color': 'darkBlue','weight': 'normal','size': 20,},pad=12)
```
